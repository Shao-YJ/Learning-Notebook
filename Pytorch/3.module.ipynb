# nn.Module

```python
from torch import nn, tensor
```

```python
class Mymodule(nn.Module):
    def __init__(self):
        super().__init__()
        
    def forward(self, x):
        return x + 1
```

```python
tudui = Mymodule()
print(tudui)
x = tensor([1, 2, 3])
x = tudui(x)
print(x)
```

## Convolution Layer

```python
import torch
import torch.nn.functional as F
```

```python
input = torch.tensor([[1, 2, 3, 4, 5],
                      [0, 1, 2, 3, 1],
                      [1, 0, 1, 0, 1],
                      [5, 2, 3, 1, 1],
                      [2, 1, 0, 1, 1]], dtype=torch.float32)
kernel = torch.tensor([[1, 0, 1],
                       [0, 1, 0],
                       [2, 0, 1]], dtype=torch.float32)

input = torch.reshape(input, (1, 1, 5, 5))
kernel = torch.reshape(kernel, (1, 1, 3, 3))

print(input.shape)
print(kernel.shape)
```

```python
output = F.conv2d(input, kernel, stride=1, padding=0)  # (5-3+1)/1=3
print(output)

output = F.conv2d(input, kernel, stride=2, padding=0)  # (5-3)/2+1=2
print(output)

output = F.conv2d(input, kernel, stride=1, padding=1)  # (5-3+2*1)/1+1=5
print(output)
```

```python
from torchvision import transforms, datasets
from torch.utils.data import DataLoader
```

```python
dataset = datasets.CIFAR10(root='./Data/cifar10', train=False, transform=transforms.ToTensor())

dataloader = DataLoader(dataset, batch_size=32, shuffle=False)

class MyConv(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 9, 3, stride=1, padding=0)  # (32-3)/1+1=30
        
    def forward(self, x):
        x = self.conv1(x)
        return x
```

```python
from torch.utils.tensorboard import SummaryWriter
```

```python
conv = MyConv()
writer = SummaryWriter("dataloader")
step = 0
for data in dataloader:
    imgs, labels = data
    output = conv(imgs)
    print(imgs.shape)
    print(output.shape)
    output = torch.reshape(output,(-1, 3, 30, 30))
    writer.add_images("input", imgs, step)
    writer.add_images("output", output, step)

    step += 1
```

## Pooling Layer

```python
torch.manual_seed(1)
```

```python
class MyPool(nn.Module):
    def __init__(self):
        super().__init__()
        self.pool1 = nn.MaxPool2d(3, ceil_mode=True)
        
    def forward(self, x):
        return self.pool1(x)
```

```python
pool = MyPool()
step = 0
writer = SummaryWriter("dataloader")
for data in dataloader:
    imgs, labels = data
    writer.add_images("pool_input", imgs, step)
    output = pool(imgs)
    writer.add_images("pool_output", output, step)
    step += 1
writer.close()
    
```

## Linear Layer

```python
class MyLinear(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = nn.Linear(32, 15)
        
    def forward(self, x):
        return self.linear1(x)
```

```python
linear = MyLinear()
step = 0
writer = SummaryWriter("dataloader")
for data in dataloader:
    imgs, labels = data
    writer.add_images("pool_input", imgs, step)
    output = linear(imgs)
    writer.add_images("pool_output", output, step)
    step += 1
writer.close()
    
```
