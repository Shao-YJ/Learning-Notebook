```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 10,
    length_function = len,
    separators = ['\n', '.', '']
)
text = "This is a sample text that will be split into smaller chunks. It contains multiple sentences and paragraphs. The text splitter will ensure that each chunk is manageable in size, while also allowing for some overlap between chunks to maintain context."

chunks = text_splitter.split_text(text)
for chunk in chunks:
    print(f"{chunk}\n")
```

```python
import time
import json
import struct
from websocket import create_connection
# 定义接收带长度字段的 WebSocket 消息的函数
def recv_message(ws):
    """
    从 WebSocket 读取带长度字段的消息。
    消息格式：
    - 前 4 字节为消息体长度（大端序）
    - 后续为实际的消息体
    """
    try:
        # 接收完整的二进制消息
        raw_data = ws.recv()
        if len(raw_data) < 4:
            raise ValueError("未收到完整的长度字段（不足 4 字节）")
        # 从前 4 字节解析消息长度
        length_data = raw_data[:4]
        message_length = struct.unpack(">I", length_data)[0]  # 使用大端序解析长度
        # print(f"接收到消息长度: {message_length}")
        # 提取消息体
        message_body = raw_data[4:]
        if len(message_body) < message_length:
            raise ValueError("未收到完整的消息体")
        return message_body
    except Exception as e:
        print(f"接收消息失败: {e}")
        return None
# WebSocket 客户端主逻辑
def websocket_client():
    # WebSocket 服务器地址
    uri = "ws://10.228.163.241:8000/ws/game/v1"
    # uri="ws://yunwei-dev.eastmoney.com/bomb-man/ws/game/v1/1"
    while True:  # 外部循环，持续尝试连接
        try:
            # 连接到 WebSocket 服务
            ws = create_connection(uri)
            print("连接到 WebSocket 服务器成功！")
            try:
                # 1. 向后端发送 "pleaseloadmap" 消息
                load_map_request = json.dumps({
                    "message": "pleaseloadmap",
                    "data": None
                })  # 构造 JSON 消息
                ws.send(load_map_request)  # 发送 JSON 消息
                print(f"已发送请求: {load_map_request}")
                # 2. 接收地图初始化数据
                init_map_data = recv_message(ws)  # 使用带长度字段的接收函数
                if init_map_data:
                    init_map_data = init_map_data.decode("utf-8")
                    print(f"收到地图初始化数据: {init_map_data}")
                # 3. 向后端发送 "pleasestartgame" 消息
                start_game_request = json.dumps({
                    "message": "pleasestartgame",
                    "data": None
                })  # 构造 JSON 消息
                ws.send(start_game_request)  # 发送 JSON 消息
                print(f"已发送请求: {start_game_request}")
                # 4. 持续接收游戏状态数据
                message_count = 0  # 初始化计数器
                while True:
                    response = recv_message(ws)  # 使用带长度字段的接收函数
                    if response:
                        response_data = json.loads(response.decode("utf-8"))  # 解析 JSON 数据
                        # 检查是否收到游戏结束指令
                        if response_data.get("message") == "pleaseendgame":
                            print("收到游戏结束指令，请停止接收数据")
                            break  # 退出内层循环
                        if response_data.get("message") == "winner":
                            print(f"赢家是: {response_data}")
                        # 增加计数器
                        message_count += 1
                        # 每 1000 条打印一次
                        if message_count % 100 == 0:
                          print(f"收到第 {message_count} 条游戏状态数据: {response_data}")
            except Exception as e:
                print(f"接收消息失败或连接中断: {e}")
            finally:
                ws.close()
                print("WebSocket 连接已关闭")
                break  # 退出外层循环
        except Exception as e:
            print(f"无法连接到 WebSocket 服务器: {e}")
            print("5 秒后重新尝试连接...")
            time.sleep(5)  # 等待 5 秒后重试连接
if __name__ == "__main__":
    websocket_client()
```
